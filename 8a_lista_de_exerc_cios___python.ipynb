## Exercício 1: Soma de 3 números

```python
# Passo 1: Solicitar ao usuário que insira os três números
numero1 = float(input("Digite o primeiro número: "))
numero2 = float(input("Digite o segundo número: "))
numero3 = float(input("Digite o terceiro número: "))

# Passo 2: Calcular a soma dos três números
soma = numero1 + numero2 + numero3

# Passo 3: Imprimir o resultado da soma
print(f"A soma dos três números é: {soma}")
```

## Exercício 2: Média aritmética de 5 números

```python
# Passo 1: Solicitar ao usuário que insira os cinco números
numero1 = float(input("Digite o primeiro número: "))
numero2 = float(input("Digite o segundo número: "))
numero3 = float(input("Digite o terceiro número: "))
numero4 = float(input("Digite o quarto número: "))
numero5 = float(input("Digite o quinto número: "))

# Passo 2: Calcular a soma dos cinco números
soma = numero1 + numero2 + numero3 + numero4 + numero5

# Passo 3: Calcular a média aritmética (soma dividida pela quantidade de números)
media = soma / 5

# Passo 4: Imprimir o resultado da média
print(f"A média aritmética dos cinco números é: {media}")
```

## Exercício 3: Média ponderada de 4 números

```python
# Passo 1: Solicitar ao usuário que insira os quatro números
numero1 = float(input("Digite o primeiro número: "))
numero2 = float(input("Digite o segundo número: "))
numero3 = float(input("Digite o terceiro número: "))
numero4 = float(input("Digite o quarto número: "))

# Passo 2: Definir os pesos correspondentes
peso1 = 1
peso2 = 2
peso3 = 3
peso4 = 4

# Passo 3: Calcular a soma dos produtos de cada número pelo seu peso
soma_produtos = (numero1 * peso1) + (numero2 * peso2) + (numero3 * peso3) + (numero4 * peso4)

# Passo 4: Calcular a soma total dos pesos
soma_pesos = peso1 + peso2 + peso3 + peso4

# Passo 5: Calcular a média ponderada (soma dos produtos dividida pela soma dos pesos)
media_ponderada = soma_produtos / soma_pesos

# Passo 6: Imprimir o resultado da média ponderada
print(f"A média ponderada dos números é: {media_ponderada}")
```

## Exercício 4: Comparar dois números

```python
# Passo 1: Solicitar ao usuário que insira os dois números
numero1 = float(input("Digite o primeiro número: "))
numero2 = float(input("Digite o segundo número: "))

# Passo 2: Comparar os números e exibir a mensagem apropriada
if numero1 > numero2:
    print("O primeiro número informado é o maior")
elif numero2 > numero1:
    print("O segundo número é o maior")
else:
    print("Ambos os números são iguais")
```

## Exercício 5: Área do triângulo

```python
# Passo 1: Solicitar ao usuário a base do triângulo
base = float(input("Digite o valor da base do triângulo: "))

# Passo 2: Solicitar ao usuário a altura do triângulo
altura = float(input("Digite o valor da altura do triângulo: "))

# Passo 3: Calcular a área do triângulo usando a fórmula Área = (Base * Altura) / 2
area = (base * altura) / 2

# Passo 4: Imprimir a área calculada
print(f"A área do triângulo é: {area}")
```

## Exercício 6: Preço final do automóvel

```python
# Passo 1: Solicitar o nome do automóvel
nome_automovel = input("Digite o nome do automóvel: ")

# Passo 2: Solicitar o preço de fábrica do automóvel
preco_fabrica = float(input("Digite o preço de fábrica do automóvel: "))

# Passo 3: Calcular o valor dos impostos (43% do preço de fábrica)
impostos = 0.43 * preco_fabrica

# Passo 4: Calcular a percentagem do revendedor (15% do preço de fábrica)
percentagem_revendedor = 0.15 * preco_fabrica

# Passo 5: Calcular o preço final do automóvel
preco_final = preco_fabrica + impostos + percentagem_revendedor

# Passo 6: Imprimir o nome do automóvel e o preço final
print(f"Nome do automóvel: {nome_automovel}")
print(f"Preço final: R$ {preco_final:.2f}")
```

## Exercício 7: Média de 15 valores reais com limite

```python
# Passo 1: Inicializar uma variável para armazenar a soma dos valores
soma_valores = 0

# Passo 2: Usar um loop para ler os 15 valores reais
for i in range(1, 16):
    valor = float(input(f"Digite o {i}º valor real: "))
    soma_valores += valor

# Passo 3: Calcular a média aritmética dos valores
media = soma_valores / 15

# Passo 4: Verificar se a média obtida é maior que 9
if media > 9:
    media_final = 10
else:
    media_final = media

# Passo 5: Imprimir a média final
print(f"A média dos valores é: {media_final:.2f}")
```

## Exercício 9: Celsius para Fahrenheit

```python
# Passo 1: Solicitar ao usuário a temperatura em Celsius
celsius = float(input("Digite a temperatura em Celsius: "))

# Passo 2: Reorganizar a fórmula C = (5/9) * (F – 32) para encontrar F:
# C * (9/5) = F - 32
# F = C * (9/5) + 32
fahrenheit = celsius * (9/5) + 32

# Passo 3: Imprimir a temperatura convertida para Fahrenheit
print(f"A temperatura de {celsius}°C corresponde a {fahrenheit:.2f}°F")
```

## Exercício 10: Quantidade de números pares de 100 a 200

```python
# Passo 1: Inicializar um contador para os números pares
contador_pares = 0

# Passo 2: Iterar pelos números de 100 a 200 (inclusive)
for numero in range(100, 201):
    # Passo 3: Verificar se o número é par (o resto da divisão por 2 é 0)
    if numero % 2 == 0:
        contador_pares += 1 # Incrementar o contador se for par

# Passo 4: Imprimir a quantidade total de números pares
print(f"A quantidade de números pares de 100 a 200 é: {contador_pares}")
```

## Exercício 11: Contar a quantidade de números pares entre dois números

```python
# Passo 1: Solicitar ao usuário os dois números
numero_inicial = int(input("Digite o primeiro número: "))
numero_final = int(input("Digite o segundo número: "))

# Passo 2: Garantir que numero_inicial seja menor ou igual a numero_final
if numero_inicial > numero_final:
    numero_inicial, numero_final = numero_final, numero_inicial # Troca os valores se estiverem invertidos

# Passo 3: Inicializar um contador para os números pares
contador_pares = 0

# Passo 4: Iterar pelos números no intervalo (inclusive)
for numero in range(numero_inicial, numero_final + 1):
    # Passo 5: Verificar se o número é par
    if numero % 2 == 0:
        contador_pares += 1

# Passo 6: Imprimir a quantidade de números pares
print(f"A quantidade de números pares entre {numero_inicial} e {numero_final} é: {contador_pares}")
```

## Exercício 12: Soma de todos os números pares entre dois números

```python
# Passo 1: Solicitar ao usuário os dois números
numero_inicial = int(input("Digite o primeiro número: "))
numero_final = int(input("Digite o segundo número: "))

# Passo 2: Garantir que numero_inicial seja menor ou igual a numero_final
if numero_inicial > numero_final:
    numero_inicial, numero_final = numero_final, numero_inicial # Troca os valores se estiverem invertidos

# Passo 3: Inicializar uma variável para armazenar a soma dos números pares
soma_pares = 0

# Passo 4: Iterar pelos números no intervalo (inclusive)
for numero in range(numero_inicial, numero_final + 1):
    # Passo 5: Verificar se o número é par
    if numero % 2 == 0:
        soma_pares += numero # Adicionar o número par à soma

# Passo 6: Imprimir a soma dos números pares
print(f"A soma dos números pares entre {numero_inicial} e {numero_final} é: {soma_pares}")
```

## Exercício 13: Fatorial de um número n

```python
# Passo 1: Solicitar ao usuário que insira um número inteiro não negativo
numero = int(input("Digite um número inteiro não negativo para calcular o fatorial: "))

# Passo 2: Verificar se o número é negativo
if numero < 0:
    print("Fatorial não existe para números negativos.")
# Passo 3: Se o número for 0, o fatorial é 1
elif numero == 0:
    print(f"O fatorial de {numero} é 1")
# Passo 4: Calcular o fatorial para números positivos
else:
    fatorial = 1
    for i in range(1, numero + 1):
        fatorial *= i # Multiplicar fatorial pelo número atual na iteração
    print(f"O fatorial de {numero} é {fatorial}")
```
